```python
def cct(t,p,q):
    while(True):
        temp_pos=0

        r=func(p,q)
        
        temp=p.copy()
        for x in r:
            if(x<t):
                y=r.index(x,temp_pos)
                temp_pos=y+1
                p.remove(y)
        if(temp==p):
            break
    return p

```
